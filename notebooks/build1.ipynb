```python
import random

names = [
    "Nguyễn Văn A", "Trần Thị B", "Lê Quốc C",
    "Phạm Thị D", "Đỗ Minh E", "Vũ Văn F"
]

addresses = [
    "Số 12 ngõ 45 đường Láng, Hà Nội",
    "123 Nguyễn Trãi, Quận 1, TP. Hồ Chí Minh",
    "45 Lê Duẩn, Quận Hải Châu, Đà Nẵng",
    "Thôn Đông, xã Tân Tiến, huyện Văn Giang, Hưng Yên"
]

phones = [
    "0987654321", "0912345678", "0905123456", "0978123456"
]

patterns = [
    ["N"], ["A"], ["P"],
    ["N", "A"], ["A", "N"], ["N", "P"], ["P", "N"],
    ["A", "P"], ["P", "A"],
    ["N", "A", "P"], ["N", "P", "A"],
    ["A", "N", "P"], ["A", "P", "N"],
    ["P", "N", "A"], ["P", "A", "N"]
]

def synthesize_entry():
    name = random.choice(names)
    addr = random.choice(addresses)
    phone = random.choice(phones)
    pattern = random.choice(patterns)
    
    parts = []
    for p in pattern:
        if p == "N":
            parts.append(name)
        elif p == "A":
            parts.append(addr)
        elif p == "P":
            parts.append("Số điện thoại " + phone)
    
    return {
        "text": ", ".join(parts),
        "pattern": "_".join(pattern)
    }

# Ví dụ sinh 5 mẫu
for _ in range(5):
    entry = synthesize_entry()
    print(entry["pattern"], "→", entry["text"])
```
