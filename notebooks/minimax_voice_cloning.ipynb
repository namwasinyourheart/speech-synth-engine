## 7. Error Handling Examples

Comprehensive error handling and edge cases:

```python
def safe_clone_with_metadata(provider, text, reference_audio, output_file):
    """
    Safe voice cloning with comprehensive error handling
    """
    try:
        print(f"ğŸ¤ Cloning: {text[:50]}...")
        
        if not reference_audio or not reference_audio.exists():
            return {'success': False, 'error': 'Reference audio not found'}
        
        if reference_audio.stat().st_size < 1000:
            return {'success': False, 'error': 'Reference audio too small'}
        
        # Clone with metadata
        result = provider.clone_with_metadata(text, reference_audio, output_file)
        
        if result.get('success') and output_file.exists():
            file_size = output_file.stat().st_size / 1024
            print(f"âœ… Success: {output_file.name} ({file_size:.1f} KB)")
            return result
        else:
            print(f"âŒ Failed: {result.get('error', 'Unknown error')}")
            return result
            
    except Exception as e:
        error_msg = f"Exception: {str(e)}"
        print(f"âŒ Error: {error_msg}")
        return {'success': False, 'error': error_msg}
    finally:
        # Always cleanup provider resources
        try:
            provider.cleanup()
        except:
            pass

# Test error handling with various scenarios
if reference_audio:
    test_cases = [
        ("Text bÃ¬nh thÆ°á»ng", "error_test_normal.wav"),
        ("", "error_test_empty.wav"),  # Empty text
        ("Text ráº¥t dÃ i " * 50, "error_test_long.wav")  # Very long text
    ]
    
    print("ğŸ§ª Testing error handling:")
    for text, filename in test_cases:
        output_file = output_dir / filename
        result = safe_clone_with_metadata(provider, text, reference_audio, output_file)
        print(f"   Result: {'âœ…' if result.get('success') else 'âŒ'}")
```
