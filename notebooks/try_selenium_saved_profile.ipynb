```python
import os
import undetected_chromedriver as uc
from selenium.webdriver.chrome.options import Options

def run_profile(profile_name):
    base_path = "/home/nampv1/selenium-profiles"
    profile_path = os.path.join(base_path, profile_name)

    # Tạo thư mục gốc nếu chưa có
    if not os.path.exists(base_path):
        os.makedirs(base_path)

    # Tạo thư mục profile nếu chưa có
    if not os.path.exists(profile_path):
        os.makedirs(profile_path)

    options = Options()
    options.add_argument(f"--user-data-dir={profile_path}")
    options.add_argument("--profile-directory=Default")

    driver = uc.Chrome(options=options)
    driver.get("https://www.minimax.io/audio/voices-cloning")

    return driver


if __name__ == "__main__":
    print("Nhập tên profile muốn dùng (vd: acc1, acc2, acc3):")
    name = input("> ")

    driver = run_profile(name)
    print(f"Đang mở profile: {name}")
    print("Lần đầu bạn cần login. Những lần sau mở lại sẽ tự vào.")
```
