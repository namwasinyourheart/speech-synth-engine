```python
import os
import sys
import logging
from pathlib import Path

# Add speech-synth-engine to path
sys.path.insert(0, "/home/nampv1/projects/tts/speech-synth-engine")

from speech_synth_engine.dataset.dataset_generator import DatasetGenerator
```

```python
from speech_synth_engine.dataset.text_loaders import (
    TextFileLoader
)
```

```python
loader = TextFileLoader("/media/nampv1/hdd/data/vn_commune_district_province/raw/text/province_list_with_prefix.txt")
texts = loader.load()
```

```python
texts
```

```python
output_dir = Path("test_output/")
output_dir.mkdir(parents=True, exist_ok=True)

# Test text tiếng Việt
test_text = "Xin chào, đây là test với Google Gemini TTS provider mới"

# Cấu hình Gemini provider
# providers_config = {
#     "gemini": {
#         "sample_rate": 24000,
#         "model": "gemini-2.5-flash-preview-tts",
#         "language": "vi"
#     }
# }

providers_config = {
    "gtts": {
        "sample_rate": 24000,
        "model": "df",
        "language": "vi"
    }
}

generator = DatasetGenerator(output_dir, providers_config)
```

```python
voice = "vi"
```

```python
results = []
```

```python
summary = generator.generate_from_text_list(
                # texts=[test_text],
                texts[:10],
                provider_model_voice_list=[("gtts", "df", voice)],
                batch_size=1,
                delay_between_requests=3.0  # Gemini cần thời gian xử lý lâu hơn
            )

if summary.successful_generations > 0:
    result = summary.results[0]
    results.append({
        'voice': voice,
        'audio_path': result.audio_path,
        'duration': result.duration
    })
```
